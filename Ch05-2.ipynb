```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np
import cv2
import matplotlib.pyplot as plt
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/gsy.jpg')
print(img.shape)
gImg1=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(gImg1.shape)
cv2_imshow(gImg1)
```

```python
outImg = cv2.blur(gImg1,(7,7))
cv2_imshow(outImg)
```

```python
outImg = cv2.GaussianBlur(gImg1,(7,7),5)
cv2_imshow(outImg)
```

```python
outImg = cv2.medianBlur(gImg1,7)
cv2_imshow(outImg)
```

```python
outImg = gImg1.copy()
blurImage = cv2.GaussianBlur(outImg,(7,7),5)
x = 290
y = 525
size = 20
for h in range(x,x+size):
  for w in range(y,y+size):
    # outImg[h][w] = 255
    outImg[h][w] = blurImage[h][w]

cv2_imshow(outImg)
```

```python
outImg = gImg1.copy()
blurImage = cv2.medianBlur(outImg, 7)
x = 290
y = 525
size = 20
for h in range(x,x+size):
  for w in range(y,y+size):
    # outImg[h][w] = 255
    outImg[h][w] = blurImage[h][w]

cv2_imshow(outImg)
```

```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/lena256.jpg')
print(img.shape)
gImg1=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(gImg1.shape)
cv2_imshow(gImg1)
```

```python
outImg = cv2.GaussianBlur(gImg1,(7,7),1)
cv2_imshow(outImg)
```

```python
edgeImage = cv2.addWeighted(gImg1,1,outImg,-1,0)
cv2_imshow(edgeImage)
```

```python
unsharpImage = cv2.addWeighted(gImg1,1,edgeImage,1,0)
cv2_imshow(unsharpImage)
```
