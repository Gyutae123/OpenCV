```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/milkdrop.bmp')
print(img.shape)
gImg2=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gImg2, cmap='gray')
```

```python
gPlus = np.zeros((256,256), dtype=np.ubyte)
```

```python
for h in range(gImg1.shape[0]):
  for w in range(gImg1.shape[1]):
    gPlus[h,w] = gImg1[h,w] + gImg2[h,w]
```

```python
plt.imshow(gPlus, cmap='gray')
```

```python
for h in range(gImg1.shape[0]):
  for w in range(gImg1.shape[1]):
    pValue = gImg1[h,w] + gImg2[h,w]
    if pValue > 255:
      gPlus[h,w] = 255
    else:
      gPlus[h,w] = pValue
plt.imshow(gPlus, cmap='gray')
```

```python
for h in range(gImg1.shape[0]):
  for w in range(gImg1.shape[1]):
    pValue = np.int32(gImg1[h,w]) + np.int32(gImg2[h,w])
    if pValue > 255:
      gPlus[h,w] = 255
    else:
      gPlus[h,w] = pValue
plt.imshow(gPlus, cmap='gray')
```

```python
for h in range(gImg1.shape[0]):
  for w in range(gImg1.shape[1]):
    gPlus[h,w] = gImg1[h,w] * 0.5 + gImg2[h,w] * 0.5

plt.imshow(gPlus, cmap='gray')
```

```python
gPlus = cv2.add(gImg1, gImg2)
plt.imshow(gPlus, cmap='gray')
```

```python
gMinus = np.zeros((gImg1.shape[0], gImg2.shape[1]), dtype=np.ubyte)
```

```python
for h in range(gImg1.shape[0]):
  for w in range(gImg1.shape[1]):
    gMinus[h,w] = gImg1[h,w] - gImg2[h,w]

plt.imshow(gMinus, cmap='gray')
```

```python
for h in range(gImg1.shape[0]):
  for w in range(gImg1.shape[1]):
    gMinus[h,w] = np.abs(np.int32(gImg1[h,w]) - np.int32(gImg2[h,w]))

plt.imshow(gMinus, cmap='gray')
```

```python
gMul = np.zeros((gImg1.shape[0], gImg2.shape[1]), dtype=np.ubyte)
```
