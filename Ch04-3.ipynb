```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np
import cv2
import matplotlib.pyplot as plt
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/girl.jpg')
print(img.shape)
gImg1=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(gImg1.shape)
cv2_imshow(gImg1)
```

```python
hist = cv2.calcHist([gImg1],[0],None,[256],[0,256])
plt.plot(hist)
```

```python
outImg = np.zeros(gImg1.shape,dtype='uint8')
```

```python
min = 0
max = 200

for h in range(gImg1.shape[0]):
  for w in range(gImg1.shape[1]):
    if gImg1[h,w] < min:
      outImg[h,w] = 0
    elif gImg1[h,w] > max:
      outImg[h,w] = 255
    else:
      val = np.int32(gImg1[h, w] - min) * 255
      val = np.int32(val / (max - min))
      outImg[h,w] = val

cv2_imshow(outImg)
```

```python
hist2 = cv2.calcHist([outImg],[0],None,[256],[0,256])
plt.plot(hist2)
```
